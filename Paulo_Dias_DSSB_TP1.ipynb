### Resolução

1. Escolha da Ferramenta de Construção:

Ambas as ferramentas são amplamente utilizadas como sistemas de construção (build tools) em projetos que utilizam linguagens da JVM. Elas automatizam tarefas recorrentes, como compilação de classes, execução de testes e empacotamento da aplicação em arquivos JAR ou WAR, poupando-nos de escrever repetidamente comandos no terminal.

**Gradle:**<br>
Oferece compilação incremental e paralela, o que resulta em builds mais ágeis. É uma ferramenta mais flexível, compatível com repositórios Maven e Ivy, de código aberto, e costuma ser a escolha mais comum em projetos mais complexos, que exigem maior customização.

**Maven:**<br>
Adota uma abordagem declarativa com etapas de build pré-definidas. É mais comum em projetos simples, com menor grau de complexidade. Sua padronização facilita a curva de aprendizado, tornando-o acessível a desenvolvedores iniciantes ou em contextos de menor exigência técnica.

**Para os fins deste projeto, optarei por utilizar o Maven como ferramenta base, considerando a natureza do projeto e seu baixo nível de complexidade.**

O processo de instalação para usuários windowns pode seguir em dois formatos:

> Spring CLI v3.5.4

> $ scoop install springboot 

 

A escolha de qual forma iniciar se dá pela necessidade do projeto, onde o Spring Boot Initializer é preferível quando se quer criar uma aplicação toda do zero, enquanto o Spring Boot CLI (Via linha de comando) é preferível quando se quer gerenciar uma aplicação já existente, para que ocorra a interação sem a necessidade de interação com uma interface web.

3. Gerenciamento de Dependências

O Spring Boot Initializr por si, possui um recurso de auto inserção de dependências (Dependências inseridas via campo "E" mensionado no exercício 1), chamado Bill of Materials (BOM), ou Lista de Materiais. Esse recurso funciona centralizando as versões de dependências, garantindo que somente versões compatíveis serão injetadas no projeto à ser desenvolvido. Ao longo do projeto, o cuidado para a injeção somente de dependências que sejam necessárias para o pleno funcionamento da aplicação, preservando assim o bom funcionamento, capacidade e desempenho da aplicação.
Para este projeto, escolhi aderir ao Lonbok pela facilidade de uso de métodos de gerenciamento (getters e setters), Spring Boot Dev Tools para funcionalidades de recarga automática e Spring Web para inclusão de configuração de servidores, como TOMCAT.

4. Utilização de Autoconfiguração

A funcionalidade de autoconfiguração do Spring Boot, permite que o próprio Spring Boot faça configurações não invasivas na sua aplicação, por exemplo Banco de Dados - adicionando MySQL ou SQLITE por exemplo, ele se encarrega das configurações de Data Source, JdbcTemplate, etc - Spring MVC - configurando o sistema com dependências como Jackson, e vem com Tomcat embutido. É possível verificar mais detalhes na [Documentação](https://docs.spring.io/spring-boot/reference/using/auto-configuration.html).<br>
Para configuração inicial, é preciso adicionar a notação @EnableAutoConfiguration ou @SpringBootApplication, ao qual o Spring Boot Initializer já se encarregou de fazer a inserção na criação do projeto.

6. Desenvolvimento de Serviços REST

Para tal, foram desenvolvidos 3 classes: > TaskFacede > TaskModels e > TaskAPI, cada uma possui suas próprias responsabilidades como descrito abaixo:

> TaskModels -> Classe responsável por determinar o modelo que será seguido, contendo como atributos, id, num1, num2 e operation; <br>
> TaskFacede -> Classe responsável por determinar os códigos que serão chamados pela API em cada caso, sendo eles métodos de soma, subtração, multiplicação, divisão, exponenciação, requisição de todo banco de dados e requisição por id; <br>
> TaskAPI -> Classe responsável pelo gerenciamento dos END POINTS, sendo eles post, getAll e getById.
